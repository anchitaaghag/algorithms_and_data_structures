````python
# TESTING VARIOUS APPROACHES

import timeit
import random

small_list = list(range(10)) + [0]*5
medium_list = list(range(10**2)) + [0]*5
big_list = list(range(10**4)) + [0]*5

random.shuffle(small_list)
random.shuffle(medium_list)
random.shuffle(big_list)

# Approach 1:

def approach1(lst):
    empty_list = []
    zeros_list = []

    for i in lst:
        if i != 0:
            empty_list.append(i)
        else:
            zeros_list.append(i)
    return empty_list + zeros_list

a = timeit.timeit("approach1(small_list.copy())", number=1000, globals=globals())
b = timeit.timeit("approach1(medium_list.copy())", number=1000, globals=globals())
c = timeit.timeit("approach1(big_list.copy())", number=1000, globals=globals())
print("Time to run on small list: {0:.6f}".format(a))
print("Time to run on medium list: {0:.6f} which is {1:.2f} times the small list".format(b, b/a))
print("Time to run on big list: {0:.6f} which is {1:.4f} times the medium list, and {1:.4f} times the small list".format(c, c/b, c/a))

# Approach 2:
# could be a remove .pop and then append the # of zeros at the end of the list

def approach2(lst):
    for i in lst:
        if i == 0:
            lst.remove(i)
            lst.append(0)
    return lst

c = timeit.timeit("approach2(small_list.copy())", number=1000, globals=globals())
d = timeit.timeit("approach2(medium_list.copy())", number=1000, globals=globals())
e = timeit.timeit("approach2(big_list.copy())", number=1000, globals=globals())
print("Time to run on small list: {0:.6f}".format(c))
print("Time to run on medium list: {0:.6f} which is {1:.2f} times the small list".format(d, d/c))
print("Time to run on big list: {0:.6f} which is {1:.4f} times the medium list, and {1:.4f} times the small list".format(e, e/d, e/c))

# Approach 3:

def approach3(lst):
    index_position = 0
    for i in lst:
        if i != 0:
            lst[index_position] = i
            index_position += 1

    for j in range(index_position, len(lst)):
        lst[j] = 0

    return lst
e = timeit.timeit("approach3(small_list.copy())", number=1000, globals=globals())
f = timeit.timeit("approach3(medium_list.copy())", number=1000, globals=globals()) 
g = timeit.timeit("approach3(big_list.copy())", number=1000, globals=globals())

print("Time to run on small list: {0:.6f}".format(e))
print("Time to run on medium list: {0:.6f} which is {1:.2f} times the small list".format(f, f/e))
print("Time to run on big list: {0:.6f} which is {1:.4f} times the medium list, and {1:.4f} times the small list".format(g, g/f, g/e))
````
