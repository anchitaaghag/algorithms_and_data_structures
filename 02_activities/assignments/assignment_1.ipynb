-   Code the solution to your assigned problem in Python (code chunk). Note: each problem can be solved more simply if you use an abstract data type that is suitable for that problem. Using that try to find the best time and space complexity solution!


````python
from typing import List

def move_zeros_to_end(nums: List[int]) -> List[int]:

    integer_list = []
    zeros_list = []

    for i in nums:
        if i != 0:
            integer_list.append(i)
        else:
            zeros_list.append(i)

    return integer_list + zeros_list

# Input lists for testing from above.

list_1 = [0, 1, 0, 3, 12]
list_2 = [4, 0, 5, 0, 0, 6]
list_3 = [1, 6, 7, 12, 2]
list_4 = [0, 0, 6, 0, 5, 0, 0, 4, 2, 0]

# Applying the user-defined function with our lists.

print(f"Example 1 Output: {move_zeros_to_end(nums=list_1)}")
print(f"Example 2 Output: {move_zeros_to_end(nums=list_2)}")
print(f"New Example 1 Output: {move_zeros_to_end(nums=list_3)}")
print(f"New Example 2 Output: {move_zeros_to_end(nums=list_4)}")
````


-   Explain why your solution works


The solution works because the code iterates through the input list once, assigns non-zero elements to the `integer_list` data structure, and assigns all zeros to the `zeros_list` data structure. Finally, the code concatenates both lists and returns the output, which ensures all zeros will be at the end of the new output list.


-   Explain the problem’s time and space complexity


The solution to this problem has a time complexity of $O(n)$ (linear time) and a space complexity of $O(n)$.

**Time Complexity: $O(n)$**

The `.append()` method is used for each individual element in the original list. This takes a constant amount of time ($O(1)$. However, since we are applying the `.append()` method for each element in the list of of size $n$, the total time taken is proportional to $n$ (i.e., the time it takes grows at the exact same speed as the number of items we start with). These operations are linear relative to the input size $n$, thus, the total time remains $O(n)$.

**Space Complexity: $O(n)$**

The function is creating two new data structures (`integer_list` and `zeros_list`) to hold the elements from the input list. As the input size $n$ increases, the total number of elements that needs to be stored in these two lists increases proportionally. When the `integer_list + zeros_list` is returned, Python creates a new list (a third list) that also contains $n$ elements. The memory usage grows lineraly with the size of the input, thus, the space complexity is $O(n)$.

**Current v.s. Alternate Solution**

The current solution would work well for small to medium sized lists. For larger lists, a for loop that manually accesses the index and assigns each element to a position may be more suitable. Performing individual index assignments thousands of times may not faster for small to medium sized lists, hence, it was not used above.

While a for loop approach has the same time complextiy as the solution above (i.e., $O(n)$ (linear time)), it has a better space complexity $O(1)$. When $n$ is large, the for loop can simply iterate over the list elements and assign the index position. In contrast, the current approach would likely need twice the space (i.e., since we are creating new duplicated lists of the original- `integer_list` and `zeros_list`).

**Alternate Solution**

We would start with creating the `def move_zeros_to_end` function that takes an list of integers called `nums`. Within this function we will create an  `index_postition=0` marker that will keep track of the moving index (i.e., postion of where we will insert the integers). Next, in our first for loop we will go through each element in `nums`, and if the element is not zero we will overwrite the current `index_position` with the value of the element. Then, we will add the index_postition with one (e.g. `+= 1`) to move to the next position.

After the first loop finishes, all the non-zero numbers are lined up at the beginning of the list, and our `index_position` variable is sitting at the very first available slot (i.e., the spot immediately following the last non-zero number).

Since we have already moved all the non-zero numbers forward, any spot from `index_position` to the end of the list should be filled with a zero. To do this, we will use a second for loop. In the second for loop, we will start at `index_position` and continue until we reach the end of the list `(len(nums))`. As the loop visits each of these remaining spots, it will assign: `nums[j] = 0`. This replaces whatever old value was sitting in that spot with a 0. Once the loop reaches the very end of the list, we will have successfully replaced all the remaining numbers with zeros. Finally, we will return the new `nums` list.
