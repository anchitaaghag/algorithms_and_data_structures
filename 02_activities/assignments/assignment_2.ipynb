  # Question Three: Missing Number in Range
 
  You are given a list containing `n` integers in the range `[1, n]`. Return a list of numbers that are missing from the range `[1, n]` of the array. If there is no missing number, return -1. Note, all the integers in the list may not be unique.
  
  ## Examples

  ### Example 1

  Input: `lst = [1, 1, 3]`

  Output: [2]

  ### Example 2

  Input: `lst = [5,3,2,4,1]`

  Output: [-1]

  ### Example 3

  Input: `lst = [6, 8, 2, 3, 5, 7, 1, 1, 2,10]`

  Output: [4, 9]

</details>

#### Starter Code for Question 3


````python
from typing import List

def missing_num(nums: List) -> int:
    # To find the missing numbers in the list, we first find the maximum number in the range
    max_num = max(nums)

    # Then, create an empty list to store the missing numbers.
    missing_numbers = []

    for i in range(1, max_num + 1):
        # If the number is in the list (including duplicates), we can skip it and move on to the next number.
        if i in nums:
            continue
        # If the number is not in the list, we can append it to a new list of missing numbers.
        else:
            missing_numbers.append(i)
    
    # If missing_numbers length == 0, return -1.
    if not missing_numbers:
        missing_numbers = [-1]
        
    return missing_numbers
````

def is_valid_brackets(s: str) -> bool:
    stack = []
    matching = {')': '(', '}': '{', ']': '['}

    for char in s:
        if char in '({[':

-   Explain why their solution works in your own words.


Both of their solutions work because they follow the LIFO (last in, first out) principle of Stack data structures. In both solutions, the function `is_valid_brackets()` drops the opening brackets on the top of the stack. Then, when the loop encounters a closing bracket, it looks at the top of the stack to see if a match is found. However, before looking at the top of the stack, the solution checks if the stack is empty (this instantly returns a `False` output). After all elements in the string are read by the function, a true/false `return len(stack) == 0` is executed which evaluates if every opening bracket found its matching closing partner (i.e., the stack is completely empty; `True`) or if there are elements still present in the stack (i.e., length is not zero; `False`).

The only difference between the two solutions, is the repetetitiveness using individual if/elif blocks in the first solution, compared to the more elegant second solution that uses a dictionary and key-value pairs.


-   Explain the problem’s time and space complexity in your own words.


The solution to this problem has a time complexity of $O(n)$ (linear time) and a space complexity of $O(n)$.

**Time Complexity: $O(n)$**

In this solution, the `is_valid_brackets()` function iterates through the input string exactly once, where $n$ is the length of the string. Inside the loop, all operations (checking characters, pushing to the stack, popping from the stack, and checking the stack length) are $O(1)$ constant-time operations. As the total time complexity is linear relative to the input size, the time complexity is $O(n)$.

**Space Complexity: $O(n)$**

As space complexity is determined assuming the worst-case scenario, a very long string that only contains opening brackets (e.g., "[[[[[[[[["), will cause Python to store all $n$ characters. Since the space complexity required to store the stack is linear relative to the input size, the space complexity is $O(n)$.

Based on Learning Support's idea for an iterative removal, I would suggest the following alternate solution:

```python

def is_valid_brackets(s: str) -> bool:
    # Keep looping as long as there is at least one adjacent matching pair to remove.
    while "()" in s or "[]" in s or "{}" in s:
        s = s.replace("()", "")
        s = s.replace("[]", "")
        s = s.replace("{}", "")
        
    # If the string is completely empty, it was valid.
    return s == ""

```
In this alternate solution, instead of tracking brackets individually using a Stack data structure, we can look at the string as a whole and continuously remove matching pairs from the inside out. If a bracket sequence is valid, there must be at least one place in the string where a matching pair sits directly side-by-side without any other characters separating them. This could be an adjacnt "()", "[]", or "{}". 

To write this user-defined function we will write a loop that runs continuously as long as the string contains at least one instance of "()", "[]", or "{}". Inside this loop, we will use a string replacement operation to find every occurrence of those three specific pairs and use `.replace()` method to substitute the matching pair with an empty string `""`. As the matching pairs disappear, the outer brackets that were separated will now move closer together, creating new adjacent pairs for the next iteration of the loop. Once the loop stops (because no more adjacent matching pairs exist), we will perform a final check. If the string has completely shrunk down to an empty string `""`, return `True`. If any characters remain behind, return `False`.

This proposed alternate solution will have the same $O(n)$ space complexity as the current Stack solution because the linear space is relative to the length of the string. However, the time complexity of this alternate solution is $O(n^2)$ (quadratic time). This is significantly less efficient than the Stack method. For every single loop iteration, python will need to scan through the entire string from left to right to find the target matching pairs, and then create a brand-new string. Performing these multiple operations inside a for loop will cause the execution time to scale quadratically with the number of characters ($n$), making it highly inefficient for large inputs compared to the single-pass linear time ($O(n)$) achieved by the Stack solution.
